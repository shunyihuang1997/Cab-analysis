```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

# EDA

```python
#function for EDA
def eda(df):
    print(df.head(2),'\n')
    print(df.info())
    print(df.describe(),'\n')
    print(df.isna().sum())
    print('Number of data point is: ', len(df))

def box_plot(df, feature):
   df.plot(y = feature, kind = 'box')

def heat_map(df):
    sns.heatmap(df.corr(),annot=True)
    
def check_vif(df,features):
    df = df.drop(features,axis = 1)
    vif_data = pd.DataFrame()
    vif_data['feature'] = df.columns
    
    vif_data['VIF'] = [variance_inflation_factor(df.values, i) for i in range(len(df.columns))]
    print(vif_data)
```

```python
#load data
#drop duplicate
#encode company and city
#check VIF
#use box plot to check outliers and heatmap for correlation
cab_df = pd.DataFrame(pd.read_csv('Cab_Data.csv'))
cab_df = cab_df.drop_duplicates()
le = LabelEncoder()
cab_df['company_encode'] = le.fit_transform(cab_df['Company'])
cab_df['city_encode'] = le.fit_transform(cab_df['City'])
cab_df['cost per km'] = cab_df['Cost of Trip']/cab_df['KM Travelled']
eda(cab_df)

check_vif(cab_df,['Transaction ID','Company','City','KM Travelled','Cost of Trip','Date of Travel'])

cab_df.plot(y = ['KM Travelled','Price Charged','Cost of Trip'],kind = 'box',subplots = True)
plt.show()
heat_map(cab_df)
plt.show()
```

```python
#load data
#remove duplicate and replace "," in the population and users
#label encode the city
#Find outliers using boxplot and use heatmap to see correlation
city_df = pd.DataFrame(pd.read_csv('City.csv'))
city_df = city_df.drop_duplicates()
city_df = city_df.replace(',','',regex = True)
le = LabelEncoder()
city_df['City_encode'] = le.fit_transform(city_df['City'])

city_df['Population'] = city_df['Population'].astype('int')
city_df['Users'] = city_df['Users'].astype('int')

eda(city_df)
city_df.plot(y = ['Population','Users'],kind = 'box', subplots=True)
plt.show()
heat_map(city_df)
plt.show()
```

```python
#load data
#remove duplicate
#encode gender
customerID_df = pd.DataFrame(pd.read_csv('Customer_ID.csv'))
customerID_df = customerID_df.drop_duplicates()
le = LabelEncoder()
customerID_df['gender_encode'] = le.fit_transform(customerID_df['Gender'])
eda(customerID_df)

check_vif(customerID_df,['Gender'])

customerID_df.plot(y = ['Age','Income (USD/Month)'], kind = 'box', subplots = True)
plt.show()
heat_map(customerID_df)
plt.show()
```

```python
#load data
#remove duplicate
#encode payment mode
transactionID_df = pd.DataFrame(pd.read_csv('Transaction_ID.csv'))
transactionID_df = transactionID_df.drop_duplicates()
le = LabelEncoder()
transactionID_df['payment_encode'] = le.fit_transform(transactionID_df['Payment_Mode'])
eda(transactionID_df)

check_vif(transactionID_df,['Payment_Mode'])

transactionID_df['Payment_Mode'].value_counts().plot(kind = 'bar',rot = 0)
plt.show()
heat_map(transactionID_df)
plt.show()
```

# Data merging

```python
transaction_left_customer = transactionID_df.merge(customerID_df, on ='Customer ID', how = 'left')
transaction_left_customer_left_cab = transaction_left_customer.merge(cab_df, on = 'Transaction ID', how = 'left')
transaction_left_customer_left_cab_left_city = transaction_left_customer_left_cab.merge(city_df, on = 'City', how = 'left')
transaction_left_customer_left_cab_left_city = transaction_left_customer_left_cab_left_city.drop(['gender_encode','company_encode','city_encode','City_encode','payment_encode'],axis = 1)
transaction_left_customer_left_cab_left_city.head(2)
```

# Visualiztion

### Hypothesis 1: Do young people (< 40 year-old) take more cab?

```python
df = pd.DataFrame(transaction_left_customer_left_cab_left_city.groupby(['Age'])['Customer ID'].count()).reset_index()
df = df.rename(columns= {'Customer ID': 'count'})
df.plot(x = 'Age',ylabel = 'total', kind = 'bar', rot = 0,cmap='Set2')
```

```python
# Most of the customers are between 18 and 40, older people might prefer to drive their own vehicle
```

### Hypothesis 2: Do big cities like New York and LA have more customers?

```python
df1 = pd.DataFrame(transaction_left_customer_left_cab_left_city['City'].value_counts()).reset_index()
df1 = df1.rename(columns={'index': 'city', 'City':'count'})
df1.plot(x = 'city',kind = 'bar',rot = 45,figsize=(25,12))
```

```python
# As we can see from the plot that big cities like New York and Chicago definitely have the most users, and smaller states/lower economic 
# states have fewer users. 
```

### Hypothesis 3: What's the most common payment by each company and how much are they

```python
df2 = pd.DataFrame(transaction_left_customer_left_cab_left_city.groupby(['Company','Payment_Mode'])['Transaction ID'].count()).reset_index()
df2 = df2.rename(columns={'Transaction ID': 'count'})
sns.barplot(x = 'Company', y ='count', data = df2, hue = 'Payment_Mode')
```

```python
# It looks like people prefer to pay using card 
```

### Hypothesis 4 : will the cost per km be more higher in big cities?

```python
df3 = pd.DataFrame(transaction_left_customer_left_cab_left_city.groupby(['City'])['cost per km'].mean()).reset_index()
df3 = df3.sort_values(by = 'cost per km', ascending = False)
df3.plot(x = 'City',ylabel = 'cost per km', kind = 'bar', rot = 45, figsize = (20,12),cmap='Set2')
```

```python
#big city cost more on gas, but the difference between the maximum and minimum is within $1
```

### Hypothesis 5: Is there any seasonality effect in the number of people using cab?

```python
df4 =  pd.DataFrame(transaction_left_customer_left_cab_left_city.groupby(['Date of Travel'])['Transaction ID'].count()).reset_index()
df4 = df4.rename(columns = {'Transaction ID':'count'})
df4.plot(x = 'Date of Travel', y = 'count')
```

```python
# The number of users do have seasonality effect, and since I don't have any idea what the date represent, I can't do deeper analysis
```

### Hypothesis 6: Number of users for each company in each city

```python
df5 = pd.DataFrame(transaction_left_customer_left_cab_left_city.groupby(['City','Company'])['Customer ID'].count()).reset_index()
df5 = df5.rename(columns={'Customer ID': 'count'})
df5 = df5.sort_values(by = 'count', ascending = False)
sns.set(rc={'figure.figsize':(20,15)})
sns.barplot(x = 'City', y ='count', data = df5, hue = 'Company')
plt.xticks(rotation=45)
plt.tight_layout()
```

```python
# In most of the cities, the number of user in yellow cab are more than the pink cab
```
